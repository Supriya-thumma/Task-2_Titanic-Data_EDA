```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
sns.set(style="whitegrid")
```

## Step 2: Loading the Dataset

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

## Step 3: Summary Statistics & Basic Info

```python
print("\nDataset Info:")
df.info()
```

```python
print("\nSummary Statistics:")
print(df.describe(include='all'))
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

## Step 4: Histograms for Numerical Features

```python
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols].hist(bins=15, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.show()
```

## Step 5: Boxplots for Numerical Features

```python
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

## Step 6: Correlation Matrix & Heatmap

```python
corr_matrix = df[num_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

## Step 7: Pairplot for Key Features

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.show()
```

## Step 8: Insights (Example)

```python
print("\nObservations from EDA:")
print("- Passengers in higher Pclass (1st class) had better survival chances.")
print("- Females had a higher survival rate than males.")
print("- Younger passengers had slightly higher survival chances.")
print("- Fare is positively correlated with survival probability.")
```
